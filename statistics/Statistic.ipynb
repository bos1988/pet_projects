## Формулы

### Параметры

Среднее арифметическое генеральной совокупности:
$$
\overline{x} = \frac{\displaystyle\sum_{i=1}^{N}x_{i}}{N}
$$

Средневзвешанная средняя генеральной совокупности:
$$
\overline{x} = \frac{\displaystyle\sum_{i=1}^{k}x_{i} N_{i}}{N}\\
$$

Среднее арифметическое выборочной совокупности:
$$
\overline{x} = \frac{\displaystyle\sum_{i=1}^{n}x_{i}}{n}
$$

Средневзвешанная средняя выборочной совокупности:
$$
\overline{x} = \frac{\displaystyle\sum_{i=1}^{k}x_{i} n_{i}}{n}\\
$$

Относительные частоты выборочной совокупности:

$$
w_{1} = \frac{n_{1}}{n},  w_{2} = \frac{n_{2}}{n}, \ldots, w_{k} = \frac{n_{k}}{n}\\
w_{1} + w_{2} + \ldots + w_{k} = 1
$$

Относительные частоты генеральной совокупности:

$$
W_{1} = \frac{N_{1}}{N},  W_{2} = \frac{N_{2}}{N}, \ldots, W_{k} = \frac{N_{k}}{N}\\
W_{1} + W_{2} + \ldots + W_{k} = 1
$$

Размах вариации:

$$
R = x_{max} - x_{min}
$$

Формула Стерджеса (оптимальное количество интервалов):

$$
k = 1 + 3.322\lg{n}
$$

Равноинтервальная группировка:

$$
h = \frac{x_{max} - x_{min}}{k} = \frac{R}{k}
$$

Плотность частот:

$$
\frac{n_{i}}{h_{i}}
$$

Плотность относительных частот:

$$
\frac{w_{i}}{h_{i}}
$$

Мода готового интервального ряда:

$$
M_{0} = x_{0}+\frac{n_{M}-n_{M-1}}{\left(n_{M}-n_{M-1}\right)+\left(n_{M}-n_{M+1}\right)}\cdot h_{M}
$$
, M - модальный интервал; $x_{0}$ - нижняя граница модального интервала

Медиана готового интервального ряда:

$$
m = x_{0}+\frac{0.5n-n_{M-1}^н}{n_{m}}\cdot h_{m}
$$
, m - медианный интервал; $x_{0}$ - нижняя граница медианного интервала, $n_{M-1}^н$ - накопленная частота предыдущего интервала.

Среднее линейное отклонение несгруппированных статистических данных:
$$
\overline{l} = \frac{\displaystyle\sum_{i=1}^{n}|x_{i} - \overline{x}|}{n}\\
$$

Среднее линейное отклонение сформированного дискретного или интервального вариационного ряда:
$$
\overline{l} = \frac{\displaystyle\sum_{i=1}^{k}|x_{i} - \overline{x}|\cdot n_{i}}{n}\\
$$

Генеральная дисперсия несгруппированных статистических данных:
$$
D = \frac{\displaystyle\sum_{i=1}^{N}\left(x_{i} - \overline{x}\right)^2}{N}\\
$$

Генеральная дисперсия сформированного дискретного или интервального вариационного ряда:
$$
D = \frac{\displaystyle\sum_{i=1}^{k}\left(x_{i} - \overline{x}\right)^2\cdot N_{i}}{N}\\
$$

Выборочная дисперсия несгруппированных статистических данных (смещённая оценка генеральной дисперсии):
$$
D = \frac{\displaystyle\sum_{i=1}^{n}\left(x_{i} - \overline{x}\right)^2}{n}\\
$$

$$
D = \overline{x^2} - \left(\overline{x}\right)^2 = \frac{\displaystyle\sum_{i=1}^{n}x_{i}^2}{n} - \Bigg(\frac{\displaystyle\sum_{i=1}^{n}x_{i}}{n}\Bigg)^2\\
$$


Выборочная дисперсия сформированного дискретного или интервального вариационного ряда (смещённая оценка генеральной дисперсии):
$$
D = \frac{\displaystyle\sum_{i=1}^{k}\left(x_{i} - \overline{x}\right)^2\cdot n_{i}}{n}\\
$$

$$
D = \overline{x^2} - \left(\overline{x}\right)^2 = \frac{\displaystyle\sum_{i=1}^{k}x_{i}^2 n_{i}}{n} - \Bigg(\frac{\displaystyle\sum_{i=1}^{k}x_{i} n_{i}}{n}\Bigg)^2\\
$$

Исправленная выборочная дисперсия (несмещённая оценка генеральной дисперсии):
$$
s^2 = \frac{n}{n-1}\cdot D
$$

Cреднее квадратическое отклонение (смещённая оценка генеральной дисперсии):
$$
\sigma = \sqrt{D}
$$

Исправленное квадратическое отклонение (несмещённая оценка генеральной дисперсии):
$$
s = \sqrt{s^2}\\
s = \sqrt{\frac{n}{n-1}}\cdot \sigma
$$

Коэффициент вариации:
$$
V = \frac{\sigma}{\overline{x}}\cdot 100\%
$$

> ***Примечание:*** *на практике часто считают именно через $\sigma$, но для оценки коэффициента вариации всей генеральной совокупности, конечно же, корректнее использовать исправленное стандартное отклонение s.*</br></br>
В статистике существует следующий эмпирический ориентир:</br></br>
– если показатель вариации составляет примерно 30% и меньше, то статистическая совокупность считается **однородной**. Это означает, что большинство вариант находится недалеко от средней, и найденное значение  хорошо характеризует центральную тенденцию совокупности.</br>
– если показатель вариации составляет существенно больше 30%, то совокупность **неоднородна**, то есть, значительное количество вариант находятся далеко от $\overline{x}$, и выборочная средняя плохо характеризует типичную варианту. В таких случаях целесообразно рассмотреть квартили, децили, а иногда и перцентили, которые делят вариационный ряд на части, и для каждого участка рассчитать свои показатели. Но это уже немного дебри статистики.

Коэффициент асимметрии:
$$
A = \frac{\overline{x}-M_{0}}{\sigma}
$$

> Из формулы следует, что в левостороннем случае коэффициент асимметрии отрицателен, а в правостороннем – положителен, и чем больше A по модулю – тем сильнее скос распределения.

Коэффициент асимметрии (продвинутая формула):
$$
A = \frac{m_{3}}{\sigma^3}
$$

Центральный эмпирический момент третьего порядка:
$$
m_{3} = \frac{\displaystyle\sum_{i=1}^{n}\left(x_{i} - \overline{x}\right)^3}{n}\\
$$

$$
m_{3} = \frac{\displaystyle\sum_{i=1}^{k}\left(x_{i} - \overline{x}\right)^3\cdot n_{i}}{n}\\
$$

> Если A > 0 , то распределение скошено вправо, если A < 0 – то влево. При этом принята следующая условная градация: если полученное значение по модулю меньше, чем 0.25, то асимметрия незначительна, если 0.25 < |A| < 0.5, то умеренная, и если |A| > 0.5, то существенная.</br>
И чем МЕНЬШЕ по модулю A, тем рассматриваемое эмпирическое распределение БЛИЖЕ к нормальному распределению с параметрами $a = \overline{x_{в}}, \sigma = \sigma_{в}$.

Коэффициент эксцесса:
$$
E = \frac{m_{4}}{\sigma^4} - 3
$$

Центральный эмпирический момент четвёртого порядка:
$$
m_{4} = \frac{\displaystyle\sum_{i=1}^{n}\left(x_{i} - \overline{x}\right)^4}{n}\\
$$

$$
m_{4} = \frac{\displaystyle\sum_{i=1}^{k}\left(x_{i} - \overline{x}\right)^4\cdot n_{i}}{n}\\
$$

> Если E > 0, то эмпирическое распределение является более высоким («островершинным») – относительно «эталонного» нормального распределения с параметрами $a = \overline{x_{в}}, \sigma = \sigma_{в}$. Если же E < 0 – то более низким и пологим. И чем больше E по модулю, тем «аномальнее» высота в ту или иную сторону.

### Доверительный интервал и доверительная вероятность (надежность)

Точность оценки:
$$
|\theta_{Г} - \theta_{в}| < \delta
$$

Доверительная вероятность (надежность интервальной оценки):

$$
P\big(|\theta_{Г} - \theta_{в}| < \delta\big) = \gamma \\
$$
Доверительный интервал:

$$
\theta_{в} - \delta < \theta_{Г} < \theta_{в} + \delta \\
(\theta_{в} - \delta; \theta_{в} + \delta)
$$

#### Генеральная средняя нормально распределённой генеральной совокупности

Доверительный интервал для оценки **генеральной средней** нормально распределённой генеральной совокупности</br>
*Если* ***известно*** *стандартное отклонение $\sigma$ генеральной совокупности.*

- Точность оценки (предельная ошибка):
$$
\delta = t_{\gamma}\cdot SE = t_{\gamma}\cdot{\frac{\sigma_{Г}}{\sqrt{n}}}
$$
- Коэффициент доверия (по функции Лапласа):
$$
2\Phi(t_{\gamma}) = \gamma
$$
- Средняя ошибка выборки:
    - для бесповторной выборки:
$$
SE = \sqrt{\frac{\sigma_{Г}^2}{n}\left(1-\frac{n}{N}\right)}
$$
    - для повторной выборки:
$$
SE = \sqrt{\frac{\sigma_{Г}^2}{n}}
$$
- Доверительный интервал:
$$
\overline{x} - t_{\gamma}\cdot{\frac{\sigma_{Г}}{\sqrt{n}}} < a < \overline{x} + t_{\gamma}\cdot{\frac{\sigma_{Г}}{\sqrt{n}}}
$$

Доверительный интервал для оценки **генеральной средней** нормально распределённой генеральной совокупности</br>
*Если генеральное стандартное отклонение $\sigma$* ***не известно***

- Точность оценки (предельная ошибка):
$$
\delta = t_{\gamma}\cdot SE = t_{\gamma}\cdot{\frac{s}{\sqrt{n}}}
$$
- Коэффициент доверия $t_{\gamma}$ рассчитывается с помощью распределения Стьюдента</br>
$k = n - 1$

- Средняя ошибка выборки:
    - для бесповторной выборки:
$$
SE = \sqrt{\frac{s^2}{n}\left(1-\frac{n}{N}\right)}
$$
    - для повторной выборки:
$$
SE = \sqrt{\frac{s^2}{n}}
$$
- Доверительный интервал:
$$
\overline{x} - t_{\gamma}\cdot{\frac{s}{\sqrt{n}}} < a < \overline{x} + t_{\gamma}\cdot{\frac{s}{\sqrt{n}}}
$$

> При большом объёме N генеральной совокупности (которая может быть и бесконечной) и малом объёма выборки n грань между формулами повторной и бесповторной выборки стирается, поскольку $\big(1-\frac{n}{N}\big)$ стремится к единице.

#### Генеральная дисперсия и стандартное отклонение

Доверительный интервал для оценки **генеральной дисперсии** и **стандартного отклонения**

*Способ первый*
$$
\frac{(n-1)s^2}{\chi_{a_{1},k}^2} < \sigma^2 < \frac{(n-1)s^2}{\chi_{a_{2},k}^2}
$$

$\chi_{a_{1},k}^2, \chi_{a_{2},k}^2$ – критические значения распределения «хи-квадрат», вычисленные для $\alpha_{1} = \frac{1-\gamma}{2}, \alpha_{2} = \frac{1+\gamma}{2} = 1-\alpha_{1}$</br>
$k = n - 1$

$$
\frac{\sqrt{(n-1)}\cdot s}{\chi_{a_{1},k}} < \sigma < \frac{\sqrt{(n-1)}\cdot s}{\chi_{a_{2},k}}
$$

*Способ второй (Как и для распределения Стьюдента, при увеличении  распределение хи-квадрат стремится к нормальному, и уже при  можно использовать приближенную формулу)*
$$
\frac{\sqrt{2n}}{\sqrt{2n-3}+t_{\gamma}}\cdot s < \sigma < \frac{\sqrt{2n}}{\sqrt{2n-3}+t_{\gamma}}\cdot s
$$
, где $2\Phi(t_{\gamma}) = \gamma$

#### Оценка генеральной доли

Генеральная доля:
$$
\omega_{Г} = \frac{K}{N}
$$
Выборочная доля:
$$
\omega = \frac{k}{n}
$$

Доверительный интервал:
$$
\omega_{в} - \Delta < \omega_{Г} < \omega_{в} + \Delta
$$

Предельная ошибка доли:
$$
\Delta = t_{\gamma}\cdot SE
$$

Средняя ошибка доли:
   - для бесповторной выборки:
$$
SE = \sqrt{\frac{\omega (1-\omega)}{n}\left(1-\frac{n}{N}\right)}
$$
   - для повторной выборки:
$$
SE = \sqrt{\frac{\omega (1-\omega)}{n}}
$$

> В том случае, если генеральная совокупность велика, а выборка малА, то для бесповторной выборки можно использовать и 2-ю формулу, ибо дробь $\displaystyle\frac{n}{N}$ будет близка к нулю.

### Статистические гипотезы

#### I) Гипотезы о числовых характеристиках статистических совокупностей, закон распределения которых уже известен.

##### Гипотеза о генеральной средней нормального распределения

Постановка задачи: предполагается, что **генеральная средняя** $a$ **нормального распределения** равна некоторому значению $a_{0}$. </br>Это нулевая гипотеза:

$H_{0}: a = a_{0}$

Для проверки гипотезы на уровне значимости $\alpha$ проводится выборка объема $n$ и рассчитывается выборочная средняя $\overline{x}_{в}$. Исходя из полученного значения и специфики той или иной задачи, можно сформулировать следующие конкурирующие гипотезы:

1) $H_{1}: a < a_{0}$ &emsp; левосторонняя область</br>
2) $H_{1}: a > a_{0}$ &emsp; правосторонняя область </br>
3) $H_{1}: a \neq a_{0}$ &emsp; двусторонняя область </br>
4) $H_{1}: a = a_{1}$, где $a_{1}$ – конкретное альтернативное значение генеральной средней.

Ошибка первого рода состоит в том, что гипотеза $H_{0}$ будет отвергнута, хотя на самом деле она правильная. Вероятность допустить такую ошибку называют уровнем значимости и обозначают буквой $\alpha$ («альфа»).  

Ошибка второго рода состоит в том, что гипотеза $H_{0}$ будет принята, но на самом деле она неправильная. Вероятность совершить эту ошибку обозначают буквой $\beta$ («бета»). Значение $1-\beta$ называют мощностью критерия – это вероятность отвержения неправильной гипотезы.

Уровень значимости задаётся исследователем самостоятельно, наиболее часто выбирают значения $\alpha = 0.1, \alpha = 0.05, \alpha = 0.01$. И тут возникает мысль, что чем меньше «альфа», тем вроде бы лучше. Но это только вроде: при уменьшении вероятности $\alpha$ - отвергнуть правильную гипотезу растёт вероятность $\beta$ - принять неверную гипотезу (при прочих равных условиях). Поэтому перед исследователем стоит задача грамотно подобрать соотношение вероятностей $\alpha$ и $\beta$, при этом учитывается тяжесть последствий, которые повлекут за собой та и другая ошибки.

Гипотеза о генеральной средней нормального распределения</br>
*Если генеральная дисперсия $\sigma$* ***известна.***

Тогда в качестве статистического критерия  рассматривают случайную величину:

$$
U = \frac{\left(\overline{X} - a_{0}\right)\sqrt{n}}{\sigma}
$$

Критическая область $u_{кр}$ определяется из соотношения:

- для односторонней гипотезы: $\displaystyle\Phi(u_{кр}) = \frac{1 - 2\alpha}{2}$


- для двусторонней гипотезы: $\displaystyle\Phi(u_{кр}) = \frac{1 - \alpha}{2}$

Наблюдаемое значение критерия:
$$
u_{набл} = \frac{\left(\overline{x}_{в} - a_{0}\right)\sqrt{n}}{\sigma}
$$

Левосторонняя критическая область:
- Если $u_{набл} > -u_{кр}$, то гипотеза $H_{0}$ на уровне значимости $\alpha$ принимается.
- Если $u_{набл} < -u_{кр}$, то отвергается.

Правосторонняя критическая область:
- Если $u_{набл} < u_{кр}$, то гипотеза $H_{0}$ на уровне значимости $\alpha$ принимается.
- Если $u_{набл} > u_{кр}$, то отвергается.

Двусторонняя критическая область:
- Если $-u_{кр} < u_{набл} < u_{кр}$ или $|u_{набл}| < u_{кр}$, то гипотеза $H_{0}$ на уровне значимости $\alpha$ принимается.
- Если $|u_{набл}| > u_{кр}$, то отвергается.


Гипотеза о генеральной средней нормального распределения</br>
*Если генеральная дисперсия $\sigma$* ***не известна.***

В этом случае остаётся ориентироваться на исправленную выборочную дисперсию  и критерий:

$$
T = \frac{\left(\overline{X} - a_{0}\right)\sqrt{n}}{s}
$$

, где $\overline{X}$ – случайное значение выборочной средней и $s$ – соответствующее исправленное стандартное отклонение.

***Данная случайная величина имеет распределение Стьюдента с $k = n - 1$ степенями свободы.***

##### Гипотеза о равенстве генеральных средних двух распределений

Постановка задачи: из двух генеральных совокупностей извлечены выборки объёмов $n$ и $m$ и найдены их выборочные средние: $\overline{x}_{в}$ и $\overline{y}_{в}$ соответственно. Требуется на уровне значимости $\alpha$ проверить гипотезу:

$H_{0}: \overline{x}_{Г} = \overline{y}_{Г}$

Конкурирующие гипотезы:

1) $H_{1}: \overline{x}_{Г} < \overline{y}_{Г}$ &emsp; левосторонняя область</br>
2) $H_{1}: \overline{x}_{Г} > \overline{y}_{Г}$ &emsp; правосторонняя область </br>
3) $H_{1}: \overline{x}_{Г} \neq \overline{y}_{Г}$ &emsp; двусторонняя область </br>

**а) независимые выборки достаточно большие $(n > 30, m > 30)$, </br>
известны генеральные дисперсии $\sigma_{x}^2$ и $\sigma_{y}^2$, </br>
генеральные совокупности распределены нормально</br>**

Статистический критерий:

$$
Z = \frac{\overline{X} - \overline{Y}} {\displaystyle\sqrt{\frac{\sigma_{x}^2}{n}+\frac{\sigma_{y}^2}{m}}}
$$

Критическая область $u_{кр}$ определяется из соотношения:

- для односторонней гипотезы: $\displaystyle\Phi(u_{кр}) = \frac{1 - 2\alpha}{2}$


- для двусторонней гипотезы: $\displaystyle\Phi(u_{кр}) = \frac{1 - \alpha}{2}$

Наблюдаемое значение критерия:
$$
z_{набл} = \frac{\overline{x}_{в} - \overline{y}_{в}} {\displaystyle\sqrt{\frac{\sigma_{x}^2}{n}+\frac{\sigma_{y}^2}{m}}}
$$

**б) независимые выборки достаточно большие $(n > 30, m > 30)$, </br>
генеральные дисперсии неизвестны, </br>
генеральные совокупности могут иметь и другое распределение (не  нормальное)</br>**

Статистический критерий:

$$
Z = \frac{\overline{X} - \overline{Y}} {\displaystyle\sqrt{\frac{D_{в}(X)}{n}+\frac{D_{в}(Y)}{m}}}
$$
, где $D_{в}(X)$ и $D_{в}(Y)$ - выборочные дисперсии 

**в) независимые малые выборки $(n < 30, m < 30)$, </br>
генеральные дисперсии неизвестны, </br>
генеральные совокупности распределены нормально</br>**

Статистический критерий:

$$
T = \frac{\overline{X} - \overline{Y}} {\sqrt{(n-1)S_{x}^2 + (m-1)S_{y}^2}} \cdot \sqrt{\frac{nm(n + m - 2)}{n + m}}
$$

, где $S_{x}^2$ и $S_{y}^2$ - исправленные выборочные дисперсии.

***Эта случайная величина распределена по закону Стьюдента с $k = n + m - 2$ степенями свободы.***

**г) зависимые выборки, </br>
генеральные дисперсии неизвестны, </br>
генеральные совокупности распределены нормально</br>**

В случае зависимых выборок гипотеза о равенстве генеральных средних сводится к уже разобранной гипотезе о значении генеральной средней и проверяется гипотеза о том, что генеральная средняя (матожидание) этой разницы равна нулю:

$H_{0}: М(\overline{D}) = 0$

Конкурирующие гипотезы:

1) $H_{1}: М(\overline{D}) < 0$ &emsp; левосторонняя область</br>
2) $H_{1}: М(\overline{D}) > 0$ &emsp; правосторонняя область </br>
3) $H_{1}: М(\overline{D}) \neq 0$ &emsp; двусторонняя область </br>

Статистический критерий:

$$
T = \frac{\left(\overline{D} - a_{0}\right)\sqrt{n}}{S} = \frac{\overline{D}\sqrt{n}}{S}
$$

, где $\overline{D} = \overline{X}_{в} - \overline{Y}_{в}$ - случайная разница между случайными значениями $\overline{X}_{в}, \overline{Y}_{в}$; </br>$S$ - исправленное стандартное отклонение разниц.

***Этот критерий имеет распределение по закону Стьюдента с количеством степеней свободы $k = n - 1$.***

##### Гипотеза о генеральной дисперсии нормального распределения

Постановка задачи: на уровне значимости $\alpha$ требуется проверить гипотезу $H_{0}: \sigma^2 = \sigma_{0}^2$ – о том, что генеральная дисперсия действительно равна своему гипотетическому значению.

Статистический критерий:

$$
\chi^2 = \frac{(n-1)S^2} {\sigma_0^2}
$$

, где $S^2$ - случайное значение исправленной дисперсии.

***Данная случайная величина имеет распределение хи-квадрат с количеством степеней свободы $k = n - 1$ и принимает лишь неотрицательные значения.***

***Распределение несимметричное, поэтому нужно оперировать левым и правым значениями в засисимости от односторонней или двусторонней гипотезы.***

##### Гипотеза о равенстве ген. дисперсий двух нормальных распределений

Постанвока задачи: из двух нормальных ген. совокупностей извлечены независимые выборки объёмом $n$ и $m$ и найдены их исправленные дисперсии: $S^2_{x}$ и $S^2_{y}$ соответственно. Требуется на уровне значимости $\alpha$ проверить гипотезу:

$H_{0}: \sigma^2_{x} = \sigma^2_{y}$


Статистический критерий:

$$
F = \frac{S^2_{Б}} {S^2_{М}}
$$

, где $S^2_{Б}$ - бОльшая исправленная дисперсия, а $S^2_{М} - меньшая.$

***Данная случайная величина имеет распределение Фишера-Снедекора (так называемое F-распределение) со степенями свободы $k_1 = n - 1, k_2 = m - 1$ если $S^2_{x} > S^2_{y}$ или $k_1 = m - 1, k_2 = n - 1$ если $S^2_{y} > S^2_{x}$. То есть, степень свободы $k_1$ соответствует выборке с бОльшей исправленной дисперсией.***

##### Гипотеза о вероятности события

Постановка задачи: в достаточно большом количестве $n$ независимых испытаний некоторое случайное событие появилось $m$ раз, и есть основание полагать, что вероятность $p$ появления этого события (в каждом испытании) равна некоторому значению $p_0$. <br/>На уровне значимости $\alpha$ требуется проверить гипотезу $H_{0}: p = p_{0}$ – о том, значимо или незначимо отличается относительная частота $\omega=\displaystyle\frac{m}{n}$ гипотетическому значению.

Статистический критерий:

$$
U = \frac{\left(\displaystyle\frac{M}{n}-p_0\right)\cdot \sqrt{n}} {\sqrt{p_oq_0}}
$$

, где $q_o=1-p_0$, a $M$ - случайное количество испытаний, в которых событие появилось. </br>При этом для качественного результата должно выполняться неравенство $np_0q_0 >\approx 10$

***Уровни значимости определяются функцией Лапласа $\Phi(u)$***

##### Сравнение вероятностей двух биномиальных распределений

Постановка задачи: в двух генеральных совокупностях проводятся независимые испытания, в каждом из которых событие $A$ может появиться – с неизвестной вероятностью $p_1$ в первой совокупности и с неизвестной вероятностью $p_2$ – во второй. По выборочным сериям испытаний объёмами $n_1$ и $n_2$ найдены соответствующие относительные частоты: $\omega_1(A)=\displaystyle\frac{m_1}{n_1}, \omega_2(A)=\displaystyle\frac{m_2}{n_2}$
, где $m_1, m_2$ – фактическое число появлений события $A$ в 1-й и во 2-й выборке. <br/>На уровне значимости $\alpha$ требуется проверить гипотезу $H_{0}: p_1 = p_2$ – о том, значимо или незначимо отличаются друг от друга относительные частоты.

Статистический критерий:

$$
U = \frac{\displaystyle\frac{M_1}{n_1} - \frac{M_2}{n_2}} {\sqrt{\displaystyle\frac{M_1+M_2}{n_1+n_2}\cdot\left(1 - \frac{M_1+M_2}{n_1+n_2}\right)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$$

, где $M_1, M_2$ - случайное количество появлений события $A$ в 1-й и во 2-й выборке соответственно

***Уровни значимости определяются функцией Лапласа $\Phi(u)$***

#### II) Гипотеза о законе распределения статистической совокупности. (Критерий согласия Пирсона.)

Требуется на уровне значимости $\alpha$ проверить нулевую гипотезу $H_0$ – о том, что генеральная совокупность распределена по закону $Z$ против конкурирующей гипотезы $H_1$ о том, что она по нему НЕ распределена.

Для ответа на вопрос **значимо или незначимо различие между эмпирическими $n_1, n_2, \dots, n_m$ и соответствующими теоретическими $n_1^t, n_2^t, \dots, n_m^t$ частотами** рассматривают различные статистические критерии, которые называют критериями согласия, и наиболее популярный из них разработал Карл Пирсон: 

$$
\chi^2 = \displaystyle\sum\frac{\left(n_i - n_i^t\right)^2}{n_i^t}
$$

При достаточно большом $n$ (объёме выборки) распределение этой **случайной величины** близкО к **распределению хи-квадрат с количеством степеней свободы $k=m-r-1$**, где $r$ – количество оцениваемых параметров закона $Z$.

Если $\chi_{набл}^2 < \chi_{кр}^2$, то на уровне значимости $\alpha$ нет оснований отвергать гипотезу $H_0$ о том, что генеральная совокупность распределена по закону $Z$.

Теоретические частоты для нормального распределения (интервальный ряд) рассчитываются по формуле:<br/>

$$
n_i^t = \displaystyle\frac{h\cdot n}{\sigma_в}\cdot f(z_i)
$$

, где $f(z)=\displaystyle\frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}$ – знакомая функция Гаусса, а $z_i = \displaystyle\frac{x_i - \overline{x_в}}{\sigma_в}$.

## Распределения:

````python
# Probability density function.
def print_pdf(distr, t:list):
    pdf_values = [distr.pdf(i) for i in t]
    print('Sigma(t)'.ljust(8), '|', 'ф(x)')
    print('-'*20)
    for i in range(len(t)):
        print(f'{t[i]:>8} | {pdf_values[i]:.6f}')
    plt.bar(t, pdf_values, width=0.2, color='steelblue')
    plt.grid(True)
    plt.show()
````

````python
#Cumulative distribution function.
def print_cdf(distr, t:list):
    cdf_values = [distr.cdf(i) for i in t]
    print('Sigma(t)'.ljust(8), '|', 'Ф(х) + 0.5'.ljust(10), '|', 'Ф(х)'.ljust(10), '|', '2Ф(х)')
    print('-'*50)
    for i in range(len(t)):
        print(f'{t[i]:>8} | {cdf_values[i]:<10.6f} | {cdf_values[i]-0.5:<10.6f} | {(cdf_values[i]-0.5)*2:.6f}')
    plt.plot(t, cdf_values, linewidth=2.5)
    plt.grid(True)
    plt.show()
````

````python
# Percent point function (inverse of cdf — percentiles).
def print_ppf(distr):
    print('Proba'.ljust(6), '|', 'Sigma')
    print('-'*20)
    for i in [0.5, 0.9, 0.95, 0.99, 0.999]:
        print(f'{i:<6} | {distr.ppf(i):.6f}')
````

````python
TITLE_FONT_SIZE = 14
LABEL_FONT_SIZE = 12
GRAPH_W = 7
GRAPH_H = 5
````

````python
# График гистограммы
def classic_hist(data:pd.Series, bins=10, title='Гистограмма', **kwargs):
    with plt.style.context('classic'):
        plt.figure(figsize=(GRAPH_W, GRAPH_H), facecolor=(0,0,0,0))
        plt.margins(0.05, 0.0)
        ax = data.hist(bins=bins, color='steelblue', **kwargs)
        plt.title(title, pad=15, fontsize = TITLE_FONT_SIZE)
        plt.xlabel('Варианта', fontsize = LABEL_FONT_SIZE)
        plt.ylabel('Количество', fontsize = LABEL_FONT_SIZE)
        y = ax.get_ybound()
        plt.ylim(y[0], y[1]*1.05)
        plt.show()
````

````python
# График гистограммы для интервального ряда
def classic_hist_for_intervals(intervals, title='Гистограмма', **kwargs):
    res = []
    for n, left, right in zip(intervals['n'], intervals['left'], intervals['right']):
        res += [left]*(n-1) + [right]
    classic_hist(pd.Series(res), bins=intervals.shape[0], title=title, **kwargs)
````

````python
# График гистограммы для интервального ряда с теоритическим распрделнием
def interval_hist(x, y, y_t, title='Гистограмма'):
    with plt.style.context('classic'):
        fig1, ax1 = plt.subplots()
        fig1.set(figwidth = GRAPH_W, figheight = GRAPH_H, facecolor = (0,0,0,0))
        plt.margins(0.05, 0.0)
        width = (max(x) - min(x)) / (len(x) - 1)
        ax1.bar(x=x, height=y, width=width, alpha=0.8, color='steelblue', edgecolor='k', linewidth = 1)
        ax1.plot(x, y_t, 'ro-', linewidth = 3)
        ax1.grid(visible=True, axis='both')
        ax1.grid(which='major',
                color = 'grey', 
                linewidth = 1.2)
        ax1.grid(which='minor', 
                color = 'grey', 
                linestyle = ':')
        ax1.set_title(title, pad=15, fontsize = TITLE_FONT_SIZE)
        ax1.set_xlabel('Варианта', fontsize = LABEL_FONT_SIZE)
        ax1.set_ylabel('Количество', fontsize = LABEL_FONT_SIZE)
        y = ax1.get_ybound()
        ax1.set_ylim(y[0], y[1]*1.05)
        plt.show()
````

````python
# График фуннкции распределения
def plot_distribution_function(intervals, title='Эмпирическая функция распределения'):
    # Добавим интервалы снизу и сверху ряда
    left_min = intervals['left'].min()
    right_max = intervals['right'].max()
    intervals = pd.concat([
        intervals, 
        pd.DataFrame([
            {'n':0, 'left':left_min-h, 'right':left_min},
            {'n':0, 'left':left_min-h*2, 'right':left_min-h},
            {'n':0, 'left':right_max, 'right':right_max+h}
        ])
    ]).sort_values('left').reset_index(drop=True)

    # Расчитаем накопленные относительные частоты
    intervals['wc'] = intervals['n'].cumsum()/intervals['n'].sum()
    
    # Построим график эмпирической функции распределения
    with plt.style.context('classic'):
        ax1 = intervals.plot(x = 'right', y = 'wc', style = 'o-', linewidth=2)
        ax1.set_title(title, pad=15, fontsize = TITLE_FONT_SIZE)
        ax1.margins(0.05, 0.05)
        ax1.figure.set(figwidth=GRAPH_W, figheight=GRAPH_H, facecolor=(0,0,0,0))
        ax1.grid(color = 'grey', linewidth = 0.5)

        ax1.set_xlabel('Варианта', fontsize = LABEL_FONT_SIZE)
        ax1.set_ylabel('Накопленная\nотносительная частота', fontsize = LABEL_FONT_SIZE)
        plt.show()
````

````python
# График полигона частот
def frequency_polygon(x, n):
    with plt.style.context('classic'):
        plt.figure(figsize=(GRAPH_W, GRAPH_H), facecolor=(0,0,0,0))
        plt.plot(x, n, 'o-', linewidth=1.5)
        plt.title('Полигон частот', pad=15, fontsize = TITLE_FONT_SIZE)
        plt.xlabel('Варианта', fontsize = LABEL_FONT_SIZE)
        plt.ylabel('Частота', fontsize = LABEL_FONT_SIZE)
        plt.grid(color = 'grey', linewidth = 0.5)
        plt.margins(0.1, 0.1)
        plt.show()
````

````python
# Функция составления интервального ряда
def make_intervals(data:pd.Series, k=10):
    # Разобъем на корзины
    intervals = pd.cut(data, bins=k).value_counts().sort_index().reset_index()
    intervals.columns = ['interval', 'n']
    # Выделим нижние и верхние границы
    intervals['left'] = intervals['interval'].apply(lambda x: round(x.left, 1))
    intervals['right'] = intervals['interval'].apply(lambda x: round(x.right, 1))
    intervals = intervals.drop('interval', axis=1)
    return intervals
````

````python
# Дополнительные вычисления для интервального ряда
def calculate_intervals(data):
    data['x'] = data.apply(lambda x: x[['left', 'right']].mean(), axis=1)
    data['w'] = data['n'] / data['n'].sum()
    data['wc'] = data['w'].cumsum()
    return data
````

#### Статистические гипотезы

````python
# Мой универсальный метод проверки гипотезы
def stat_test(*args, side='two', alpha=0.05, scipy_distr=st.t, criterion=0):
    c = {'l': (1, 0), 'r': (0, 1), 't': (0.5, 0.5), 'd': (0.5, 0.5), '2': (0.5, 0.5)}[str(side)[0]]
    a_left = c[0] * alpha
    a_right = 1 - c[1] * alpha
    
    pvalue = scipy_distr.cdf(criterion, *args)
    u_left = scipy_distr.ppf(a_left, *args)
    u_right = scipy_distr.ppf(a_right, *args)
    cr = [u_left, criterion, u_right]
    cr_str = [f'( {u_left:.6f} ', f' \x1b[1m{criterion:.6f}\x1b[0m ', f' {u_right:.6f} )']
    cr_ind = sorted(range(3), key = lambda x: cr[x])
    a_str = [f'( {a_left:.4f} ', f' \x1b[1m{pvalue:.4f}\x1b[0m ', f' {a_right:.4f} )']
    
    print('\nДиапазон критерия:')
    [print(cr_str[cr_ind[i]], end=' ') for i in range(3)]
    print('\n\nДиапазон верояности:')
    [print(a_str[cr_ind[i]], end=' ') for i in range(3)]
    
    print('\n\nНулевая гипотеза', end=' ')
    if a_left < pvalue < a_right:
        print('\x1b[1mне отвергается\x1b[0m')
    else:
        print('\x1b[1mотвергается\x1b[0m')
````

##### Гипотеза о законе распределения статистической совокупности. (Критерий согласия Пирсона.)

На уровне значимости 0,05 проверить гипотезу о том, что генеральная совокупность распределена нормально. Построить эмпирическую гистограмму и теоретическую кривую.

````python
alpha = 0.05
````

Выборочную среднюю $\overline{x_в}$ и выборочное стандартное отклонение $\sigma_в$ мы уже нашли:

````python
print(f'Выборочная средняя = {a}')
print(f'Выборочное стандартное отклонение = {sigma}')
````

Эмпирические частоты известны из предложенного интервального ряда, и осталось найти теоретические:

Теоретические частоты для нормального распределения (интервальный ряд) рассчитываются по формуле:<br/>

$$
n_i^t = \displaystyle\frac{h\cdot n}{\sigma_в}\cdot f(z_i)
$$

, где $f(z)=\displaystyle\frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}$ – знакомая функция Гаусса, а $z_i = \displaystyle\frac{x_i - \overline{x_в}}{\sigma_в}$.

````python
h_discrete = (discrete['x'].max() - discrete['x'].min()) / (discrete['x'].count() - 1)
print('Длина интервала дискретного ряда =', h_discrete)
````

````python
discrete['w_t'] = discrete['x'].apply(lambda x: st.norm.pdf(x, loc=a, scale=sigma))
````

````python
discrete['n_t'] = discrete['w_t'] * discrete['n'].sum() * h_discrete #/ sigma_i
````

*На сигма можно не делить, поскольку `st.norm.pdf()` сама это делает при использовании `loc=a_i, scale=sigma_i`*

````python
discrete
````

````python
interval_hist(discrete['x'], discrete['n'], discrete['n_t'], title='Гистограмма частот')
````

````python
print(f'Объем выборки - {n}')
print(f'Сумма теоритических частот - {discrete["n_t"].sum()}')
````

Это объяснимо тем, что эмпирическая гистограмма конечна, а нормальная кривая – бесконечна, и небольшой «недобор» теоретических частот приходится на участки, лежащие слева и справа от гистограммы. 

*1 способ (по теории)*

*Только правостороняя гипотеза*

$$
\chi^2 = \displaystyle\sum\frac{\left(n_i - n_i^t\right)^2}{n_i^t}
$$

При достаточно большом $n$ (объёме выборки) распределение этой **случайной величины** близкО к **распределению хи-квадрат с количеством степеней свободы $k=m-r-1$**, где $r$ – количество оцениваемых параметров закона $Z$.

````python
chi_right = st.chi2.ppf(1-alpha, discrete['x'].count() - 2 - 1)
print(f'Значение правого критического критерия: {chi_right}')
````

````python
u = ((discrete['n'] - discrete['n_t'])**2 / discrete['n_t']).sum()
print(f'Значение наблюдаемого критерия: {u}')
````

````python
print('Нулевая гипотеза', end=' ')
if u < chi_right:
    print('не отвергается')
else:
    print('отвергается')
````

````python
# Мода интервального ряда
def get_mode(intervals):
    mode_index = advanced_intervals['n'].idxmax()
    x0 = advanced_intervals.loc[mode_index, 'left']
    n = advanced_intervals.loc[mode_index, 'n']
    n_prev = 0 if mode_index == 0 else advanced_intervals.loc[mode_index-1, 'n']
    n_next = 0 if mode_index == len(advanced_intervals)-1 else advanced_intervals.loc[mode_index+1, 'n']
    M0 = x0 + (n - n_prev) / (n * 2 - n_prev - n_next) * h
    print(f'Модальный интервал: {advanced_intervals.loc[mode_index, ["left", "right"]].values} с частотой {n}')
    return M0
````

Эмпирические частоты известны из предложенного интервального ряда, и осталось найти теоретические:

Теоретические частоты для нормального распределения (интервальный ряд) рассчитываются по формуле:<br/>

$$
n_i^t = \displaystyle\frac{h\cdot n}{\sigma_в}\cdot f(z_i)
$$

, где $f(z)=\displaystyle\frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}$ – знакомая функция Гаусса, а $z_i = \displaystyle\frac{x_i - \overline{x_в}}{\sigma_в}$.

Это объяснимо тем, что эмпирическая гистограмма конечна, а нормальная кривая – бесконечна, и небольшой «недобор» теоретических частот приходится на участки, лежащие слева и справа от гистограммы.

*Только правостороняя гипотеза*

$$
\chi^2 = \displaystyle\sum\frac{\left(n_i - n_i^t\right)^2}{n_i^t}
$$

При достаточно большом $n$ (объёме выборки) распределение этой **случайной величины** близкО к **распределению хи-квадрат с количеством степеней свободы $k=m-r-1$**, где $r$ – количество оцениваемых параметров закона $Z$.

### Биноминальное распределение

Относительная частота события (статистическая вероятность события):

$$
\omega = \frac{m}{n}
$$

Доверительный интервал и доверительная вероятность

Точность оценки:
$$
|\omega - p| < \varepsilon
$$

Доверительная вероятность (надежность интервальной оценки):

$$
P\big(|\omega - p| < \varepsilon \big) = \gamma
$$

Доверительный интервал:
$$
\omega - \varepsilon < p < \omega + \varepsilon \\
(\omega - \varepsilon; \omega + \varepsilon)
$$

Точность оценки:

$$
\varepsilon = t_{\gamma}\cdot \sqrt{\frac{\omega (1-\omega)}{n}}
$$
Коэффициент доверия (по функции Лапласа):
$$
\gamma = 2\Phi\big(t_{\gamma}\big) = 2\Phi\Bigg(\frac{\varepsilon n}{\sigma}\Bigg) = 2\Phi\Bigg(\frac{\varepsilon n}{\sqrt{pqn}}\Bigg) = 2\Phi\Bigg(\frac{\varepsilon \sqrt{n}}{\sqrt{\omega (1-\omega)}}\Bigg)
$$
Доверительный интервал:
$$
\omega - t_{\gamma}\cdot \sqrt{\frac{\omega (1-\omega)}{n}} < p < \omega + t_{\gamma}\cdot \sqrt{\frac{\omega (1-\omega)}{n}}
$$

p - Генеральная доля

#### Примеры:

Проводят независимые испытания с одинаковой, но неизвестной вероятностью p появления события A в каждом испытании. Найти доверительный интервал для оценки вероятности p с надёжностью $\gamma = 0.95$, если в n=300 испытаниях событие A появилось m=75 раз.

````python
m = 75
n = 300
w = m / n
print(f'Относительная частота события w = {w}')
````

````python
ty = st.norm.interval(0.95)[1]
print(f'Коэффициент доверия t = {ty}')
````

````python
delta = ty * (w*(1-w)/n)**0.5
print(f'Точность оценки d = {delta}')
````

````python
print(f'Доверительный интервал:\n{w-delta:.4f} < p < {w+delta:.4f}')
````

Из 500 поступивших на сортировку шариков для подшипников 200 попало в первую группу. В предположении о биномиальном распределении, определить:

1) доверительную вероятность того, что найденная доля шариков отклонится от вероятности попадания шарика в первую группу, менее чем на 0,03.
2) доверительную вероятность того, что вероятность попадания шарика в 1-ю группу будет накрыта интервалом (0.35; 0.5).

1) 

````python
m = 200
n = 500
w = m / n
print(f'Относительная частота события w = {w}')
````

````python
e = 0.03
ty = e * (n/w/(1-w))**0.5
print(f'Коэффициент доверия t = {ty}')
print(f'Надежность y = {(st.norm.cdf(ty) - 0.5)*2}')
````

2)

````python
e = w - 0.35
ty = e * (n/w/(1-w))**0.5
print(f'Коэффициент доверия t1 = {ty}')
Phi1 = (st.norm.cdf(ty) - 0.5)
print(f'Ф(t1) = {Phi1}')
e = 0.5 - w
ty = e * (n/w/(1-w))**0.5
print(f'Коэффициент доверия t = {ty}')
Phi2 = (st.norm.cdf(ty) - 0.5)
print(f'Ф(t2) = {Phi2}')
print()
print(f'Надежность y = {Phi1 + Phi2}')
````

Проверив 1000 изделий, обнаружили, что 300 изделий первого сорта. Сколько надо проверить изделий, чтобы с уверенностью 95% определить долю первосортных изделий с точностью до 0,01?

````python
e = 0.01
m = 300
n = 1000
w = m / n
print(f'Относительная частота события w = {w}')
````

````python
ty = st.norm.interval(0.95)[1]
print(f'Коэффициент доверия t = {ty}')
````

````python
n = (ty / e)**2 * w * (1-w)
print(f'Чтобы с уверенностью 95% определить долю первого сорта с точностью до 0,01, нужно проверить {math.ceil(n)} изделий')
````
